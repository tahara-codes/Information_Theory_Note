```python
import numpy as np
import matplotlib.pyplot as plt

# エントロピーを計算する関数
def calculate_entropy(p):
    # エラー処理
    if p == 0 or p == 1:
        return 0
    # コインのエントロピーを計算
    else:
      return -p * np.log2(p) - (1 - p) * np.log2(1 - p)

# 確率pの値を0から1まで0.01刻みで生成
p_list = np.linspace(0.0, 1.0, 100)

# 各pに対するエントロピーを計算
entropy_list = [calculate_entropy(p) for p in p_list]

# グラフの描画
plt.plot(p_list, entropy_list, '-b')
plt.xlabel('Probability of Heads (p)')
plt.ylabel('Entropy (bits)')
plt.legend()
plt.grid(True)
plt.show()
```
